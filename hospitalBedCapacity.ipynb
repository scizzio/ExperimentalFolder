```python
x = intersectData['Percentage of Available ICU Beds Needed, Twelve Months-2']
y = intersectData['Percentage of Potentially Available ICU Beds Needed, Twelve Months-2']

# Create a DataFrame with the desired columns
data = pd.DataFrame({'Percentage of Available ICU Beds Needed, Twelve Months-2': x, 'Percentage of Potentially Available ICU Beds Needed, Twelve Months-2': y})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plot the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```
